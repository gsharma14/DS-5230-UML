We extracted the loss values for the generators and discriminators that were trained in both the CycleGAN models and the CUT model. We plot these values for each epoch

```python
# general packages
import pandas as pd
import numpy as np
from scipy import stats
import pickle
import re
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
sns.set()
```

```python
# saved loss results per epoch
cycle_gan = pd.read_csv('cycle_gan_loss.csv')
cycle_gan_aug = pd.read_csv('cycle_gan_aug_loss.csv')
cut = pd.read_csv('CUT_loss.csv')

print(cycle_gan.shape)
print(cycle_gan_aug.shape)
print(cut.shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4))

ax[0].plot(cycle_gan_aug['EPOCH'], cycle_gan_aug['MONET_GEN_LOSS'], 'b-')
ax[0].plot(cycle_gan_aug['EPOCH'], cycle_gan_aug['PHOTO_GEN_LOSS'], 'g-')
ax[0].legend(['Monet-G Loss', 'Photo-G Loss'])
ax[0].set_xlabel('Epoch', fontsize=10)
ax[0].set_ylabel('Loss', fontsize=10)
ax[0].set_title('Generator Loss', fontsize=10)


ax[1].plot(cycle_gan_aug['EPOCH'], cycle_gan_aug['MONET_DISC_LOSS'], '-', color="orange")
ax[1].plot(cycle_gan_aug['EPOCH'], cycle_gan_aug['PHOTO_DISC_LOSS'], 'r')
ax[1].legend(['Monet-D Loss', 'Photo-D Loss'])
ax[1].set_xlabel('Epoch', fontsize=10)
ax[1].set_ylabel('Loss', fontsize=10)
ax[1].set_title('Discriminator Loss', fontsize=10)

fig.suptitle('Cycle GAN (Augmented) Loss', fontsize=12) 
fig.tight_layout()
fig.subplots_adjust(top=0.88);
```
